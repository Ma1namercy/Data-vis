```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

Bar graph with Matplotlib

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]

# Create a bar plot
plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

Using Seaborn to create Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a random 4x4 matrix
data = np.random.rand(5, 5)

# Create a heatmap
sns.heatmap(data, annot=True, cmap="Blues")
plt.title("Heatmap Example")
plt.show()
```

Creating Box plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create a box plot
sns.boxplot(x="day", y="total_bill", data=data)
plt.title("Box Plot Example")
plt.show()
```

PLOTLY

Creating  interactive line

```python
import plotly.express as px

# Sample data
df = px.data.gapminder()

# Create an interactive line plot
fig = px.line(df, x="year", y="pop", color="continent", title="Population Over Time")
fig.show()
```

Piechart

```python
import plotly.graph_objects as go

# Data
labels = ['A', 'B', 'C', 'D']
values = [10, 20, 30, 40]

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.update_layout(title="Pie Chart Example")
fig.show()
```
